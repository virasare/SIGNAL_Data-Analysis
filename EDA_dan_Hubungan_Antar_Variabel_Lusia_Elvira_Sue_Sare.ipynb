```python
import pandas as pd

df = pd.read_csv("Survei Pengalaman dan Kepuasan Pengguna Aplikasi Samsat Digital Nasional (SIGNAL) (Jawaban) - Form Responses 1.csv")
df.head()
```

```python
# Melihat jumlah baris dan kolom
print(f"Shape dataset: {df.shape}")
```

```python
df.info()  # Melihat tipe data tiap kolom dan jumlah data non-null
```

```python
df.describe()  # Statistik deskriptif untuk kolom numerik
```

```python
df.isnull().sum()  # Mengecek apakah ada data yang hilang
```

```python

# Menampilkan jumlah nilai unik dalam setiap kolom
print(df.nunique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df["Umur"], bins=30, kde=True)
plt.show()
```

```python
# Mengganti nama pada kolom pertanyaan, sehingga tidak terlalu panjang dan mudah dipahami

df = df.rename(columns={
    "Aplikasi mudah digunakan dan tidak membingungkan.": "Mudah Digunakan",
    "Proses pembayaran pajak kendaraan melalui aplikasi berjalan dengan lancar.": "Pembayaran Lancar",
    "Saya merasa aman saat memasukkan data pribadi di aplikasi ini.": "Keamanan Data",
    "Informasi yang diberikan dalam aplikasi akurat dan mudah dipahami.": "Info Akurat",
    "Waktu yang dibutuhkan untuk menyelesaikan transaksi pajak kendaraan cukup cepat.": "Transaksi Cepat",
    "Tampilan dan desain aplikasi menarik serta nyaman digunakan.": "Desain Menarik",
    "Aplikasi memiliki fitur yang lengkap dan sesuai dengan kebutuhan saya.": "Fitur Lengkap",
    "Aplikasi memiliki fitur yang lengkap dan sesuai dengan kebutuhan saya..1": "Fitur Lengkap 2",  # Ada duplikasi, sesuaikan jika perlu
    "Saya merasa puas dengan layanan yang diberikan oleh aplikasi ini.": "Kepuasan Layanan",
    "Saya akan merekomendasikan aplikasi ini kepada orang lain.": "Rekomendasi",
    "Saya berharap aplikasi ini terus dikembangkan agar semakin baik.": "Harapan Pengembangan"
})

# Buat Heatmap Setelah Mengganti Nama Kolom
correlation_matrix = df.select_dtypes(include=["number"]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.scatterplot(x=df["Umur"], y=df["Mudah Digunakan"])
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Perbesar ukuran gambar agar lebih jelas
sns.boxplot(x=df["Jenis Kelamin"], y=df["Kepuasan Layanan"])
plt.xticks(rotation=45)  # Putar tulisan 45 derajat
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df.select_dtypes(include=['number']))
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="Umur", data=df)
plt.xticks(rotation=45)
plt.show()
```
